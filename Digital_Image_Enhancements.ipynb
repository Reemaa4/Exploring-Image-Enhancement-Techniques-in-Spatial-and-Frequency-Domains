```python
import os
import cv2
import numpy as np
import matplotlib.pyplot as plt

# Dataset path
image_folder = r"PNEUMONIA"

# Load the images from image_folder
images = []
for filename in os.listdir(image_folder):
    img_path = os.path.join(image_folder, filename)
    img = cv2.imread(img_path)

    if img is not None:
            images.append(img)

# Display the number of images loaded
print(f"Loaded {len(images)} images from the dataset.")
```

```python
# Applying Two Spatial Domain Enhancement Techniques For The Dataset:

for img in images:

    img = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

    # 1. Histogram Equalization
    equalized_img = cv2.equalizeHist(img)
    
    # 2. Image Smoothing
    blurred_img = cv2.GaussianBlur(img, (5, 5), 0)

    # Display the results
    plt.figure(figsize=(12, 6))
    plt.subplot(131), plt.imshow(img, cmap='gray'), plt.title('Original Image')
    plt.subplot(132), plt.imshow(equalized_img, cmap='gray'), plt.title('Histogram Equalization')
    plt.subplot(133), plt.imshow(blurred_img, cmap='gray'), plt.title('Image Smoothing')
    plt.show()
```
